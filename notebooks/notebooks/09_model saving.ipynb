# Phase 9: Model Saving

**Project:** AI Chatbot for College Enquiry System  
**Phase:** 9 - Model Saving

## Objective

As per project requirements, we must save the trained model and supporting files using **Joblib** (recommended for scikit-learn models) so that they can be loaded in the Streamlit application.

**Files to be Saved:**
- Trained Model
- TF-IDF Vectorizer
- Label Encoder (if used)

```python
import joblib
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# ================== YOUR CUSTOM PATHS ==================
MODELS_PATH = r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\models'
# =====================================================

# Create models folder if not exists
os.makedirs(MODELS_PATH, exist_ok=True)

print("✅ Models folder ready at:", MODELS_PATH)
```

```python
import joblib

# ============================================
# LOAD TRAINED MODEL
# ============================================

best_model = joblib.load(
    r"C:\Users\hemanjali\Desktop\clg enquiry chatbot\models\best_model.pkl"
)

# ============================================
# LOAD TF-IDF VECTORIZER
# ============================================

vectorizer = joblib.load(
    r"C:\Users\hemanjali\Desktop\clg enquiry chatbot\models\tfidf_vectorizer.pkl"
)

print("✅ Model Loaded Successfully!")

print("✅ Vectorizer Loaded Successfully!")
```

```python
# 1. Save the Best Model
joblib.dump(best_model, f'{MODELS_PATH}/chatbot_model.pkl')

# 2. Save TF-IDF Vectorizer
joblib.dump(vectorizer, f'{MODELS_PATH}/tfidf_vectorizer.pkl')

# 3. Create and Save Label Encoder
df = pd.read_csv('data/processed_data.csv')
le = LabelEncoder()
le.fit(df['tag'])
joblib.dump(le, f'{MODELS_PATH}/label_encoder.pkl')

print("✅ All Files Saved Successfully in Models Folder!")
```

```python
# Verify saved files
saved_files = os.listdir(MODELS_PATH)
print("📁 Files Saved in Models Folder:\n")
for file in saved_files:
    print("   •", file)
```

```python
summary = {
    'File Name': ['chatbot_model.pkl', 'tfidf_vectorizer.pkl', 'label_encoder.pkl'],
    'Purpose': [
        'Contains the trained Optimized Logistic Regression model',
        'Converts text queries into numerical features',
        'Converts intent names to numbers and vice versa'
    ],
    'Saved Path': [f'{MODELS_PATH}/chatbot_model.pkl']*3
}

summary_df = pd.DataFrame(summary)
display(summary_df)
```

## Model Saving Summary

**Method Used:** `joblib` (Better than pickle for large ML models)

**Saved Files:**

1. **`chatbot_model.pkl`** → Final Trained Model
2. **`tfidf_vectorizer.pkl`** → Text preprocessing object
3. **`label_encoder.pkl`** → Converts between intent names and numbers

These files will be used in **Phase 10 (Streamlit Application)** to make predictions.
