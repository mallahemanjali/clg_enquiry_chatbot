# Phase 5: Train-Test Split

**Project:** AI Chatbot for College Enquiry System

## Why Train-Test Split is Required?

Train-Test Split is a crucial step in Machine Learning for the following reasons:

1. **To evaluate model performance** on unseen data.
2. **To prevent overfitting** — the model should not only memorize training data but also generalize well.
3. **To get a realistic estimate** of how the model will perform in real-world scenarios.
4. **To avoid data leakage** between training and testing phases.

Without train-test split, we cannot reliably measure the model's accuracy and generalization capability.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os

print("✅ Libraries Imported Successfully!\n")

# Load the processed dataset
df = pd.read_csv('data/processed_data.csv')

print("✅ Data Loaded Successfully!")
print("Original Dataset Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
# Define Features (X) and Target (y)
X = df['pattern']      # User queries (Input)
y = df['tag']          # Intent (Target Variable)

# Perform Train-Test Split (80:20)
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.20,          # 20% for testing
    random_state=42,         # For reproducibility
    stratify=y               # Maintain class balance
)

print("✅ Train-Test Split Completed Successfully!")
print("="*50)
print("Training Samples  :", X_train.shape[0])
print("Testing Samples   :", X_test.shape[0])
print("Training Ratio    :", round((X_train.shape[0] / len(X)) * 100, 2), "%")
print("Testing Ratio     :", round((X_test.shape[0] / len(X)) * 100, 2), "%")
print("="*50)
```

## Split Ratio Used

**Ratio: 80 : 20**

- **80% Training Data** → Used to train the model
- **20% Testing Data**  → Used to evaluate the model

**Why 80:20?**
- Best balance between training and evaluation for medium-sized datasets.
- Sufficient data for model to learn patterns.
- Enough test data to get reliable accuracy score.

```python
# Create folder to save split data
os.makedirs('data/split', exist_ok=True)

# Save all splits
X_train.to_csv('data/split/X_train.csv', index=False)
X_test.to_csv('data/split/X_test.csv', index=False)
y_train.to_csv('data/split/y_train.csv', index=False)
y_test.to_csv('data/split/y_test.csv', index=False)

print("✅ All Train-Test split files saved successfully in 'data/split' folder!")
```

## Summary of Phase 5

- **Method Used**: `train_test_split()` from scikit-learn
- **Ratio**: 80:20
- **Stratified Split**: Applied (to keep same proportion of each intent in train and test)
- **Random State**: 42 (ensures same split every time)

This split prepares our data for Model Building in the next phase.
