# Phase 2: Data Collection

**Project:** AI Chatbot for College Enquiry System

## 1. Dataset Source

- **Dataset Name:** College Chatbot Intents Dataset (Intent)
- **Platform:** Kaggle
- **Link:** [https://www.kaggle.com/datasets/niraliivaghani/chatbot-dataset](https://www.kaggle.com/datasets/niraliivaghani/chatbot-dataset)
- **File Name:** `intents.json`
- **Format:** JSON
- **Reason for Selection:** This dataset contains real college-related queries with proper intent tags, making it highly suitable for our project.

## 2. Dataset Features Explanation

| Column      | Description                                      | Type          | Example Value                     |
|-------------|--------------------------------------------------|---------------|-----------------------------------|
| `tag`       | Intent category (Target Variable)                | String        | `admission`, `fees`, `greeting`   |
| `pattern`   | Example user queries (Input Feature)             | String        | "What is the fee structure?"      |
| `response`  | Bot's reply for that intent                      | String        | "The B.Tech fee is ₹1,25,000..."  |

- **Total Patterns:** Around 300+ (varies)
- **Total Intents:** 35–40

```python
import json
import pandas as pd
import os

# ================== YOUR FULL PATH ==================
JSON_PATH = r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\data\intents.json'
# ====================================================

print("Loading JSON file...")

with open(JSON_PATH, 'r', encoding='utf-8') as f:
    data = json.load(f)

print("✅ JSON Loaded! Total Intents:", len(data['intents']))

# Convert to DataFrame
rows = []
for intent in data['intents']:
    for pattern in intent['patterns']:
        response = intent['responses'][0] if intent.get('responses') else "Sorry, I don't understand."
        rows.append({
            'tag': intent['tag'],
            'pattern': pattern,
            'response': response
        })

df = pd.DataFrame(rows)

print("\nDataset Shape:", df.shape)
print("Unique Intents:", df['tag'].nunique())

# Create data folder and save CSV
os.makedirs(r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\data', exist_ok=True)
df.to_csv(r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\data\/processed_data.csv', index=False)

print("\n🎉 SUCCESS! processed_data.csv has been created.")
print("File Location: data/processed_data.csv")
```

## Feature Meanings

1. **tag** → This is the **Target Variable**. The model will learn to predict this from the user's question.

2. **pattern** → These are training examples of what users might ask. Used as input features (X).

3. **response** → Predefined answers that the chatbot will give when a particular intent is detected.
