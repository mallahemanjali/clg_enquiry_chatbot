# Phase 3: Exploratory Data Analysis (EDA)

**Project:** AI Chatbot for College Enquiry System

## Data Understanding

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ================== YOUR FULL PATH ==================
DATA_PATH = r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\data\processed_data.csv'
# ====================================================

df = pd.read_csv(DATA_PATH)

print("=== Data Understanding ===")
print("1. Shape of Dataset:", df.shape)
print("\n2. Columns and Data Types:")
print(df.info())
print("\n3. Statistical Summary:")
print(df.describe(include='all'))
print("\n4. Missing Values:")
print(df.isnull().sum())
print("\n5. Duplicate Records:", df.duplicated().sum())
```

## Visualization

```python
plt.figure(figsize=(12, 6))
sns.countplot(y=df['tag'], order=df['tag'].value_counts().index)
plt.title('Distribution of Intents (Count Plot)')
plt.xlabel('Count')
plt.ylabel('Intent')
plt.tight_layout()
plt.savefig(r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\screenshots\eda\intent_distribution.png')
plt.show()
```

```python
# Create text length column
df['text_length'] = df['pattern'].apply(len)

plt.figure(figsize=(10, 6))
sns.histplot(df['text_length'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Query Text Length')
plt.xlabel('Length of Query')
plt.ylabel('Frequency')
plt.savefig(r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\screenshots\eda\intent_distribution.png')
plt.show()

print("Average Query Length:", df['text_length'].mean())
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['text_length'])
plt.title('Box Plot of Query Length')
plt.savefig(r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\screenshots\eda\intent_distribution.png')
plt.show()
```

```python
df['word_count'] = df['pattern'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 6))
sns.histplot(df['word_count'], kde=True, bins=20, color='green')
plt.title('Distribution of Word Count in Queries')
plt.xlabel('Number of Words')
plt.savefig(r'C:\Users\hemanjali\Desktop\clg enquiry chatbot\screenshots\eda\intent_distribution.png')
plt.show()
```
