```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
import os

MODELS_PATH = r"C:\Users\hemanjali\Desktop\clg enquiry chatbot\notebooks\notebooks\models"

df = pd.read_csv('data/processed_data.csv')

le = LabelEncoder()
le.fit(df['tag'])   # Fit on ALL intents

os.makedirs(MODELS_PATH, exist_ok=True)
joblib.dump(le, f'{MODELS_PATH}/label_encoder.pkl')

print("✅ Label Encoder Fixed!")
print("Total Intents:", len(le.classes_))
```
