```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries Imported!")
```

```python
# Load data
df = pd.read_csv(r"C:\Users\hemanjali\Desktop\clg enquiry chatbot\data\processed_data.csv")

X = df['pattern']
y = df['tag']

# Improved TF-IDF Vectorization
vectorizer = TfidfVectorizer(
    ngram_range=(1, 3),      # Unigrams + Bigrams + Trigrams
    max_df=0.85,
    min_df=2,
    max_features=5000
)

X_tfidf = vectorizer.fit_transform(X)

print("Feature Matrix Shape:", X_tfidf.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, 
    test_size=0.20, 
    random_state=42, 
    stratify=y
)

print("Training Shape:", X_train.shape)
print("Testing Shape :", X_test.shape)
```

```python
# Hyperparameter Tuning for Logistic Regression (Fast & Effective)
param_grid = {
    'C': [0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs'],
    'max_iter': [2000]
}

base_model = LogisticRegression()

grid_search = GridSearchCV(
    base_model, 
    param_grid, 
    cv=5, 
    scoring='accuracy', 
    n_jobs=-1,
    verbose=1
)

print("🔄 Starting Hyperparameter Tuning...")
grid_search.fit(X_train, y_train)

print("\n✅ Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)

# Best Model
best_model = grid_search.best_estimator_
```

```python
# Cross Validation Score
cv_scores = cross_val_score(best_model, X_tfidf, y, cv=5)
print("Cross Validation Scores:", cv_scores)
print("Mean CV Accuracy:", cv_scores.mean())
```

```python
import joblib

# Save Best Model
joblib.dump(
    best_model,
    r"C:\Users\hemanjali\Desktop\clg enquiry chatbot\models\best_model.pkl"
)

# Save Vectorizer
joblib.dump(
    vectorizer,
    r"C:\Users\hemanjali\Desktop\clg enquiry chatbot\models\tfidf_vectorizer.pkl"
)

print("✅ Files Saved Successfully!")
```
